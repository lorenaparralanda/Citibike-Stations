```python
import fiona
import geopandas as gpd
from geopandas import GeoDataFrame 
import shapefile
import pandas as pd
import csv
```

```python
#BRING IN NYC BLOCKS
blocs = gpd.GeoDataFrame.from_file('mappy.shp')
blocs.head()
```

```python
blocs.shape
```

```python
#CALCULATE BLOCKS CENTROIDS
blocs['centroid'] = blocs.geometry.centroid
```

```python
#BRING IN ZONES 
zones = gpd.GeoDataFrame.from_file('Zoning GIS Data- Shapefile.geojson')
zones.head()
```

```python
#CALCULATE ZONES CENTROIDS
zones['centroid'] = zones.geometry.centroid
```

```python
zones.head()
```

```python
zones.shape
```

```python
# LOCATE THE ZONE EACH BLOCK BELONGS TO
def bloctozone(x):
    distancebloctozone = []
    y = zones.geometry
    for i,j in enumerate(y):           
        distance= x.distance(j)
        distancebloctozone.append((distance, i))
    return min(distancebloctozone)
```

```python
blocs['bloctozone_dist_index'] = blocs.centroid.apply(bloctozone)
```

```python
blocs.head()
```

```python
blocs.to_csv("blocs&zones.csv")
```

```python
blocs_zones = blocs
```

```python
# CLEAN INFORMATION
blocs_zones['closest_zone_distance'] = blocs_zones.bloctozone_dist_index.apply(lambda x: x[0])
blocs_zones['closest_zone_index'] = blocs_zones.bloctozone_dist_index.apply(lambda x: x[1])
blocs_zones['closest_zone_id'] = blocs_zones.closest_zone_index.apply(lambda x: zones.loc[x]['overlay'])
```

```python
blocs_zones.head()
```

```python
blocs_zones = blocs_zones.rename(columns = {'closest_zone_id':'zone_type'})
```

```python
blocs_zones.to_csv("15.Blocs_&_zones.csv")
```
