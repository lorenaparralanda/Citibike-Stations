# ORGANIZE STATION DATA (SQL)

```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
cnx = create_engine('postgresql://username:password@ip:port/database')
```

## 1. Created trip table

```python
# CREATED TABLE (COPIED INFO INTO TABLE)
CREATE TABLE trips (id SERIAL PRIMARY KEY, tripduration INT, starttime TIMESTAMP, stoptime TIMESTAMP, startstationid INT,  
startstationname text, startstationlatitude DOUBLE PRECISION, startstationlongitude DOUBLE PRECISION, 
endstationid INT, endstationname TEXT, endstationlatitude DOUBLE PRECISION, endstationlongitude DOUBLE PRECISION, 
bikeid INT, usertype TEXT, birthyear INT, gender INT);
```

```python
# TOTAL TRIPS FROM JULY 2013 TO DECEMBER 2016
pd.read_sql_query('''SELECT COUNT(*) FROM trips;''',cnx)
```

```python
# NUMBER OF UNIQUE START STATIONS
pd.read_sql_query('''SELECT COUNT(DISTINCT(startstationname)) AS NumStations FROM trips''',cnx)
```

```python
# NUMBER OF UNIQUE END STATIONS
pd.read_sql_query('''SELECT endstationame, COUNT(DISTINCT(endstationname)) AS NumStations FROM trips''',cnx)
```

```python
# FIRST 5 LINES OF TRIP DATA
pd.read_sql_query('''SELECT * FROM trips LIMIT 5''',cnx)
```

```python
# MOST USED START STATION
pd.read_sql_query('''SELECT startstationname, COUNT(*) AS stationcount FROM trips GROUP BY startstationname ORDER BY stationcount DESC LIMIT 50''',cnx)
```

```python
# MOST USED END STATION
pd.read_sql_query('''SELECT endstationname, COUNT(*) AS stationcount FROM trips GROUP BY endstationname ORDER BY stationcount DESC LIMIT 50''',cnx)
```

## 3. Joined station birth information with lat/long

```python
# JOINED STATION NAME, BIRTH WITH LOCATION
stationpoints = pd.read_sql_query('''select distinct(o.startstationname),t.startstationlatitude,t.startstationlongitude,t.startstationid,o.datetime,o.year,o.month from orderdbirth as o join trips as t on o.startstationname=t.startstationname;''',cnx)
```

```python
# SAVED to CSV
stationpoints.to_csv('stationpoints.csv')
```

```python
# CALCULATE LIFE OF STATION IN MONTHS
import csv
months = pd.read_csv("stationpoints.csv")
```

```python
months.head()
```

```python
months['months_left'] = 12 - months.month
months['years_in_as_months'] = (2016 - months.year)*12
months['total_months'] = abs(months.years_in_as_months + months.months_left)
```

```python
months = months.drop(['months_left', 'years_in_as_months','Unnamed: 0'], 1)
```

```python
months.head()
```

```python
# CHECKING OUTLIERS
months[months['total_months']==0]
```

```python
# FIXING OUTLIERS
months.set_value(48, 'total_months', 1)
months.set_value(128, 'total_months', 1)
```

```python
import csv
months.to_csv("station_loc_birth.csv") 
```

## 4. Calculate station averages

```python
# CLEAN UP STATION INFO
months = months.rename(columns = {'startstationid':'stationid'})
```

```python
months = months.rename(columns = {'startstationname':'stationname'})
months = months.rename(columns = {'startstationlatitude':'stationlat'})
months = months.rename(columns = {'startstationlongitude':'stationlong'})
```

```python
months.head()
```

```python
months.shape
```

```python
# NUMBER OF TRIPS PER START STATION FROM JULY 2013 TO DECEMBER 2016
totaltrips = pd.read_sql_query('''select startstationid, count(*) from trips group by startstationid;''', cnx)
```

```python
totaltrips = totaltrips.rename(columns = {'count':'startcount'})
```

```python
totaltrips.head()
```

```python
totaltrips.shape
```

```python
# NUMBER OF TRIPS PER END STATION FROM JULY 2013 TO DECEMBER 2016
totalendtrips = pd.read_sql_query('''select endstationid, count(*) from trips group by endstationid;''', cnx)
```

```python
totalendtrips = totalendtrips.rename(columns = {'count':'endcount'})
```

```python
totalendtrips.head()
```

```python
totalendtrips.shape
```

```python
months.head()
```

```python
# MERGE STATION TO STATION AS START STATION INFORMATION
stationinfo = pd.merge(months, totaltrips, how='left', left_on="stationid", right_on='startstationid')
```

```python
stationinfo.head()
```

```python
# MERGE STATION TO STATION AS END STATION INFORMATION
stationinfo = pd.merge(stationinfo, totalendtrips, how='left', left_on="stationid", right_on='endstationid')
```

```python
stationinfo.head()
```

```python
stationinfo = stationinfo.drop(['startstationid', 'endstationid'],1)
```

```python
# CALCULATE AVERAGES
stationinfo['as_start_avg'] = stationinfo.startcount/stationinfo.total_months
stationinfo['as_end_avg'] = stationinfo.endcount/stationinfo.total_months
```

```python
# CLEAN UP NAMES
stationinfo = stationinfo.rename(columns = {'startcount':'as_start_count'})
stationinfo = stationinfo.rename(columns = {'endcount':'as_end_count'})
```
